```python
# Sample data: Monthly sales
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [15000, 18000, 16500, 19200, 17800, 20000]

plt.plot(months, sales, marker='o', linewidth=2, markersize=8, color='#2E86AB')
plt.title('Monthly Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales ($)')

# Add value labels on points
for i, value in enumerate(sales):
    plt.text(i, value + 300, f'${value:,}', ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

# Bar Chart

```python
df_sales = pd.read_csv('../datasets/sales_data.csv')
df_sales
```

```python
region_sales = df_sales.groupby('region')['total_amount'].sum().reset_index()
region_sales.plot(kind='bar', x='region', y='total_amount', color='#2E86AB', legend=False)

plt.title('Total Sales by Region', fontsize=16, fontweight='bold')
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)

plt.xticks(rotation=45)

plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
```

```python
region_sales = df_sales.groupby('region')['total_amount'].sum().reset_index()
region_sales.plot(kind='barh', x='region', y='total_amount', color='#2E86AB', legend=False)

plt.title('Total Sales by Region', fontsize=16, fontweight='bold')
plt.ylabel('Region', fontsize=12)
plt.xlabel('Total Sales ($)', fontsize=12)

plt.xticks(rotation=45)

plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
```

```python
region_sales = df_sales.groupby('region')['total_amount'].sum().reset_index()
sns.barplot(data=region_sales, x='region', y='total_amount', palette='viridis')

plt.title('Total Sales by Region', fontsize=16, fontweight='bold')
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)

plt.xticks(rotation=45)

plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
```

# Pie Chart

```python
payment_counts = df_sales['payment_method'].value_counts()
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Sales Distribution by Payment Method', fontsize=16, fontweight='bold')
plt.axis('equal')
plt.show()
```

```python
df_products = pd.read_csv('../datasets/product_catalog.csv')

# Create color map for categories
categories = df_products['category'].unique()
colors = sns.color_palette('husl', n_colors=len(categories))
category_colors = dict(zip(categories, colors))

# Plot each category with different color
for category in categories:
    mask = df_products['category'] == category
    plt.scatter(df_products[mask]['cost_price'], 
                df_products[mask]['retail_price'], 
                alpha=0.6, 
                s=100, 
                c=[category_colors[category]], 
                edgecolors='black', 
                linewidth=0.5,
                label=category)

plt.title('Retail Price vs Cost Price by Category', fontsize=16, fontweight='bold')
plt.xlabel('Cost Price ($)', fontsize=12)
plt.ylabel('Retail Price ($)', fontsize=12)

# Add a diagonal reference line (where retail = cost)
max_val = max(df_products['cost_price'].max(), df_products['retail_price'].max())
plt.plot([0, max_val], [0, max_val], 'r--', alpha=0.5, label='Break-even line', linewidth=2)

plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better-looking plots
sns.set_style("whitegrid")

# Create figure and axis
plt.figure(figsize=(12, 8))

# Create scatter plot
sns.scatterplot(data=df_products, 
                x='cost_price', 
                y='retail_price', 
                hue='category',
                s=100,  # marker size
                alpha=0.7,  # transparency
                edgecolor='white',
                linewidth=0.5)

# Customize the plot
plt.title('Retail Price vs Cost Price by Category', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Cost Price', fontsize=12, fontweight='bold')
plt.ylabel('Retail Price', fontsize=12, fontweight='bold')

# Improve legend
plt.legend(title='Category', title_fontsize=12, fontsize=10, 
           loc='upper left', frameon=True, shadow=True)

# Add a diagonal reference line (where retail = cost)
ax = plt.gca()
lims = [
    max(ax.get_xlim()[0], ax.get_ylim()[0]),
    min(ax.get_xlim()[1], ax.get_ylim()[1])
]
plt.plot(lims, lims, 'k--', alpha=0.3, zorder=0, label='Break-even line')

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df_sales['total_amount'], bins=30, color='#2E86AB', edgecolor='black', alpha=0.7)

# Add average line
avg_amount = df_sales['total_amount'].mean()
plt.axvline(avg_amount, color='red', linestyle='--', linewidth=2, label=f'Average: ${avg_amount:.2f}')

plt.title('Distribution of Total Sales Amount', fontsize=16, fontweight='bold')
plt.xlabel('Total Amount ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Create bins with equal distance of 100
min_val = df_sales['total_amount'].min()
max_val = df_sales['total_amount'].max()
bins = np.arange(min_val, max_val + 100, 100)

plt.hist(df_sales['total_amount'], bins=bins, color='#2E86AB', edgecolor='black', alpha=0.7)

# Add average line
avg_amount = df_sales['total_amount'].mean()
plt.axvline(avg_amount, color='red', linestyle='--', linewidth=2, label=f'Average: ${avg_amount:.2f}')

plt.title('Distribution of Total Sales Amount', fontsize=16, fontweight='bold')
plt.xlabel('Total Amount ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

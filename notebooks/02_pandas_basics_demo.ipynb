# Filtering Data

```python
# Filter: Sales above $1000
df_sales[df_sales['total_amount']>1000]
```

```python
# Filter: Sales from the North region
df_sales[df_sales['region'] == 'North']
```

```python
# Filter: AND condition
df_sales[(df_sales['region'] == 'North') & (df_sales['total_amount']>500)]
```

```python
# Filter: OR condition
df_sales[(df_sales['region'] == 'North') | (df_sales['total_amount']>500)]
```

```python
# Filter: Sales from the North region
df_sales[df_sales['region'].isin(['North', 'South'])]
```

# Sorting Data

```python
df_sales.sort_values(by='total_amount', ascending=True)
```

```python
df_sales.sort_values(by=['payment_method','total_amount'], ascending=True)
```

# Basic Calculations

```python
# total sales
total_sales = df_sales['total_amount'].sum()
total_sales
```

```python
df_sales['payment_method'].unique()
# df_sales['payment_method'].nunique()
df_sales['payment_method'].value_counts()
df_sales['payment_method'].value_counts(normalize=True)
```

# Adding new columns

```python
df_sales['discount_10pct'] = df_sales['total_amount']*0.1
df_sales['amount_after_discount'] = df_sales['total_amount'] - df_sales['discount_10pct']
df_sales['amount_after_discount']
```

# Working with dates

```python
df_sales['date'] = pd.to_datetime(df_sales['date'])
df_sales['year'] = df_sales['date'].dt.year
df_sales['month'] = df_sales['date'].dt.month
df_sales['day'] = df_sales['date'].dt.day
df_sales['day_of_week'] = df_sales['date'].dt.day_name()
df_sales[['date','year','month','day','day_of_week']]
```

# Practice Exercise: Customer data analysis

```python
import pandas as pd
df_customers = pd.read_csv("../datasets/customer_data.csv")
```

```python
df_customers.info()
```

```python
# Exercise 1: Find premium customers
df_customers[df_customers['customer_segment'].str.lower() == 'premium']
```

```python
df_customers.sort_values(by='lifetime_value', ascending=False).head(10)
```

```python
df_reset = df_customers.nlargest(10, 'lifetime_value').reset_index(drop=True)
df_reset
```

```python
# Exercise 3: Customer statistics by segment
for i in df_customers['customer_segment'].unique():
    df_customer_segment = df_customers[df_customers['customer_segment'] == i]
    count = len(df_customer_segment)
    avg_value = df_customer_segment['lifetime_value'].mean()
    print(f"{i}: {count} customers, Avg lifetime value: ${avg_value:,.2f}")
```
